```python
import os
import datetime

# os.listdir()
```

```python
matriz = []
with open('tp8_lista.csv') as archivo:
    for row in archivo:
        row.replace(' ','')
        row.replace('\t','')
        fila = row.split(',')[:-1]
        fila[0] = datetime.datetime.strptime(fila[0], '%Y-%m-%d')
        fila[1] = float(fila[1])
        matriz.append(fila)
c0=[]
c1=[]
for fila in matriz:
    c0.append(fila[0])
    c1.append(fila[1])
fig = plt.figure(figsize=(10,3))
plt.plot(c0,c1,'--xr')
plt.show()
```
